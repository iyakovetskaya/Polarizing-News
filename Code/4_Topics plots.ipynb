```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
topic_num_name=pd.read_csv('../Data/LDA_final_topic_names.csv', index_col=0)[['New_short']].rename(
    columns={'New_short': 'Topic'})
summary_table=pd.read_csv('../Data/Model/LDA_final_40_topics_summary_table.csv', index_col=0)
summary_table=summary_table.sort_index()
```

## Figure 3

```python
coef=pd.read_csv('../Data/reg_1_coef.csv')
coef=coef.iloc[51:, :].reset_index(drop=True)

y = summary_table['Relevance'].to_list()[:-1]
y = np.array(y)*100
z = coef['x'].to_list()
n = topic_num_name['Topic'].to_list()[:-1]

y_s = summary_table['Relevance'].to_list()[-1]*100

fig, ax = plt.subplots(figsize=(16,12))
ax.scatter(z, y, color='tab:blue') 
ax.set_xlim(-30,35)
ax.set_ylim(0,7)

for i, txt in enumerate(n):
    if i in [22, 7, 30, 5, 10, 31, 29, 33, 9, 37]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]-0.12),  fontsize=12)
    elif i in [19]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-0.65, y[i]+0.05),  fontsize=12)
    else:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]+0.05),  fontsize=12)

ax.scatter([0], [y_s], color='tab:blue')
ax.annotate('Social Media', (0, y_s), (-1, y_s+0.1),  fontsize=12)
        
ax.set_ylabel('Prevalence (%)')
ax.set_xlabel('Estimate')
ax.set_title('Topic Prevalence vs. Regression Coefficient (OLS1)', fontsize=16, pad=10)  
#fig.tight_layout()
plt.subplots_adjust()
plt.show()
```

## Change over time

```python
coef1=pd.read_csv('../Data/sep_reg_before_coef.csv')
coef2=pd.read_csv('../Data/sep_reg_after_coef.csv')

coef1=coef1.iloc[45:, :].reset_index(drop=True)
coef2=coef2.iloc[46:, :].reset_index(drop=True)
len(coef1['x'].to_list())
```

### Before November 2020 (Figure A7a)

```python
y = summary_table['Relevance'].to_list()[:-1]
y = np.array(y)*100
z = coef1['x'].to_list()
n = topic_num_name['Topic'].to_list()[:-1]

y_s = summary_table['Relevance'].to_list()[-1]*100

fig, ax = plt.subplots(figsize=(16,12))
ax.scatter(z, y, color='tab:blue') 
ax.scatter([0], [y_s], color='tab:blue')
ax.set_xlim(-37,27)
ax.set_ylim(0,7)

for i, txt in enumerate(n):
    if i in [13, 30, 19, 5, 22, 8, 25]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]-0.1),  fontsize=12)
    elif i in [14]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-5.5, y[i]+0.05),  fontsize=12)
    else:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]+0.05),  fontsize=12)

ax.annotate('Social Media', (0, y_s), (0.05, y_s+0.05),  fontsize=12)

ax.set_ylabel('Prevalence (%)')
ax.set_xlabel('Estimate')
ax.set_title('Topic Prevalence vs. Regression Coefficient < Elections', fontsize=16, pad=10)  
#fig.tight_layout()
plt.subplots_adjust()
plt.show()
```

```python
y = summary_table['Relevance'].to_list()[:-1]
y = np.array(y)*100
z = coef2['x'].to_list()
n = topic_num_name['Topic'].to_list()[:-1]

y_s = summary_table['Relevance'].to_list()[-1]*100

fig, ax = plt.subplots(figsize=(16,12))
ax.scatter(z, y, color='tab:blue')
ax.scatter([0], [y_s], color='tab:blue')
ax.set_xlim(-37,27)
ax.set_ylim(0,7)

for i, txt in enumerate(n):
    if i in [7, 19, 5, 22, 31, 17]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]-0.1),  fontsize=12)
    elif i in [8]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-3.2, y[i]-0.1),  fontsize=12)
    elif i in [37]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-5.7, y[i]+0.05),  fontsize=12)
    elif i in [30]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]-0.15),  fontsize=12)
    else:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]+0.05),  fontsize=12)

        
ax.annotate('Social Media', (0, y_s), (0.05, y_s+0.05),  fontsize=12)

ax.set_ylabel('Prevalence (%)')
ax.set_xlabel('Estimate')
ax.set_title('Topic Prevalence vs. Regression Coefficient > Elections', fontsize=16, pad=10)  
#fig.tight_layout()
plt.subplots_adjust()
plt.show()
```

### Figure 4

```python
z = coef1['x'].to_list()
y = coef2['x'].to_list()
n = topic_num_name['Topic'].to_list()[:-1]
n[24]='China, India'
n[17]='Sports'

fig, ax = plt.subplots(figsize=(16,14))
ax.scatter(z, y, color='tab:blue') #s=[a*5000 for a in y]
#ax.plot([0,0], [-4,4], linestyle='dashed', color='tab:blue')
ax.set_xlim(-35,30)
ax.set_ylim(-35,25)

for i, txt in enumerate(n):
    if i in [2,5,6,14,21,29,30,35]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]-1),  fontsize=12)
    elif i in [37]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.05, y[i]+0.7),  fontsize=12)
    elif i in [7, 32]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-6, y[i]+0.2),  fontsize=12)
    elif i in [9]:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]-6.8, y[i]+0.2),  fontsize=12)
    else:
        ax.annotate(' '.join([a.capitalize() for a in txt.split(' ')]), (z[i], y[i]), (z[i]+0.1, y[i]+0.2),  fontsize=12)

ax.scatter([0], [0], color='tab:blue')
ax.annotate('Social Media', (0, 0), (0.05, 0.1),  fontsize=12)


import numpy as np
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(np.array(z).reshape((-1, 1)), np.array(y))
ax.plot([-100,100],[model.intercept_-model.coef_[0]*100,model.intercept_+model.coef_[0]*100], 
        color='tab:red', linestyle='dashed', alpha=0.5)

ax.set_ylabel('Prevalence (%)')
ax.set_xlabel('Estimate Pre Election')
ax.set_ylabel('Estimate Post Election')
ax.set_title('<elections and >elections coefs comparision', fontsize=16, pad=10)  
#fig.tight_layout()
plt.subplots_adjust()
plt.show()
```
